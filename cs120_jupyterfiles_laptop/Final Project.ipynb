```python
import checkbook as c
from IPython.display import clear_output
def runbas():
    '''
    Runs the interface / deals with the user input from calculations done by checkbook.py
    '''
    r = 1
    while r == 1:
        print("Welcome to your digital Checkbook.")
        print("If this is your first use, please choose option: reset(0).")
        print("Choose 1 from the following:")
        print("  (1) Check balance")
        print("  (2) Withdrawl")
        print("  (3) Deposit")
        print("  (4) Transactions from date")
        print("  (5) Withdrawl & Deposit")
        print("  (6) Major Withdralws")
        print("  (7) Major Deposits")
        print("  (8) Print checkbook")
        print("  (9) Exit")
        print("  (0) Reset")
        decision = int(input("Your choice -> "))
        if decision == 1:
            clear_output(wait=False)
            print("Current Balance:",c.balance())
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 2:
            clear_output(wait=True)
            date = str(input("Date (mm/dd/yy) ->"))
            note = str(input("Note ->"))
            withdrawl = float(input("Withdrawl amt ->"))
            c.writeline(date,note,withdrawl,0.0)
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 3:
            clear_output(wait=True)
            date = str(input("Date (mm/dd/yy) ->"))
            note = str(input("Note ->"))
            deposit = float(input("Deposit amt ->"))
            c.writeline(date,note,0,deposit)
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 4:
            clear_output(wait=True)
            print("Enter below to find all transactions from date")
            x = str(input("(mm/dd/yy) ->"))
            c.dateTransactions(x)
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 5:
            clear_output(wait=True)
            date = str(input("Date (mm/dd/yy) ->"))
            note = str(input("Note ->"))
            withdrawl = float(input("Withdrawl amt ->"))
            deposit = float(input("Deposit amt ->"))
            c.writeline(date,note,withdrawl,deposit)
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 6:
            clear_output(wait=True)
            x = float(input("Find all withdrawls over amount () ->"))
            c.majorWithdrawls(x)
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 7:
            clear_output(wait=True)
            x = float(input("Find all deposits over amount () ->"))
            c.majorDeposits(x)
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 8:
            clear_output(wait=True)
            c.outputLog()
            print("checkbook created.")
            print("checkbook can be found under 'chck_output.txt'.")
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        if decision == 9:
            clear_output(wait=True)
            print("Goodbye.")
            r = -1
        if decision == 0:
            clear_output(wait=True)
            e1 = str(input("Would you like to reset? y/n ->"))
            if e1 == "y":
                e2 = str(input("Are you sure? y/n ->"))
                if e2 == "y":
                    print("Application reset.")
                    c.setup()
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        elif decision > 9 or decision <0:
            clear_output(wait=True)
            print("Invalid.")
            exit = str(input("Return to main menu? y/n ->"))
            if exit == "n":
                print("Goodbye.")
                r = -1
            clear_output(wait=True)
        
```

```python
runbas()
```
