The following functions are all intended to check whether a string contains any lower case letters, but at least some of them wrong. For each function, describe what the function actually does (assuming the variable s is a string), and explain why.

```python
def any_lowercase1(s):
  for c in s:
        if c.islower():
            return True
        else:
            return False
 ##any_lowercase1() takes an argument (a word) and checks to see if any
 ##characters are lower case. If at least one letter is lowercase,
 ##the function returns the bool True. If not, it returns False.   
```

```python
def any_lowercase2(s):
    for c in s:
        if 'c'.islower():
            return 'True'
        else:
            return 'False'
##any_lowercase2() will always return the string "True". The function
##checks to see if the character 'c' is lowercase, in which it always is,
##so it always returns the string "True".
```

```python
def any_lowercase3(s):
    for c in s:
        flag = c.islower()
    return flag
 ##any_lowercase3() takes an argument (a word) and checks to see if any
 ##characters are lower case. If at least one letter is lowercase,
 ##the function sets the bool variable "flag" to True and returns it. 
 ###If not, the function returns False.   
```

```python
def any_lowercase4(s):
    flag = False
    for c in s:
        flag = flag or c.islower()
    return flag
###The bool "flag" has an initial attribute of "False."
##within the for-loop, "flag" gets set to the value 
##"flag" (already false) or set to True, if c.islower() finds
##a lowercase letter. Flag then is returned as "True" if there
#is a lowercase letter or "False" if there is not.
```

```python
def any_lowercase5(s):
    for c in s:
        if not c.islower():
            return False
    return True
##within the for-loop, the if-statement checks to see if 
##there are any characters that equal c.islower(). The function will always
##return true unless no lower case character is found. For it to return False,
##The if-statement needs to find zero lower case letters; then it executes.
```
