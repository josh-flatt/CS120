# Homework 8 - Dictionaries

A permutation is a map that gives a reordering of some items. For example the permutation:  'a' -> 'b', 'b'->'c', 'c'->'a' maps a to b, b to c, and c to d.

We can represent a permutation as a dictionary where the key is the item being mapped and the value is the item it is mapped to.

## 1 - There are 6 permutations of the three letters 'a', 'b', and 'c'.  Define the six dictionaries we could use to represent these.

Note that one of the six permutations is the one that just sense 'a' -> 'a', 'b' -> 'b', and 'c'->'c'. We call this the identity permutation.

```python
#Part 1
perm_dict1 = dict()
perm_dict1['a'] = 'a'
perm_dict1['b'] = 'b'
perm_dict1['c'] = 'c'

perm_dict2 = dict()
perm_dict2['a'] = 'b'
perm_dict2['b'] = 'c'
perm_dict2['c'] = 'a'

perm_dict3 = dict()
perm_dict3['a'] = 'c'
perm_dict3['b'] = 'a'
perm_dict3['c'] = 'b'

perm_dict4 = dict()
perm_dict4['a'] = 'c'
perm_dict4['b'] = 'b'
perm_dict4['c'] = 'a'

perm_dict5 = dict()
perm_dict5['a'] = 'b'
perm_dict5['b'] = 'a'
perm_dict5['c'] = 'c'

perm_dict6 = dict()
perm_dict6['a'] = 'a'
perm_dict6['b'] = 'c'
perm_dict6['c'] = 'b'
```

```python
#Part 2
def prod(dict1,dict2):
  prod_out = dict()
  for x in dict1:
    tempval1 = dict1[x]
    tempval2 = dict2[tempval1]
    prod_out[x]=tempval2
  return prod_out
```
