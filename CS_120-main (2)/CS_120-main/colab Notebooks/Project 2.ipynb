```python

####################Reads URL Input, Nicely Outputs 10 most common words and amount used######################

import urllib.request
breakers = [' ','!','"','#','$','~','%','&','(',')','*','+',',','.','/',':',';','<','=','>','?','@','[',']','_']
  #-> list of characters that signify the end or beginning of a word; also used to omit special characters
  #     from the word count.
tempwordct = []
  #Holds all words found in url
orderset = []
  #Holds each unique word; order of most used in descending order
wordct = dict()
  #Holds each unique word and correlating word count; not in order
sort_d = dict()
  #Holds each unique word and correlating word count; in descending order of times used

######################################################
def output():
  '''
Neatly prints the 10 most common words found in website (url)
uses list(orderset) to print the actual word in place of dictionary keys.
uses dict(sort_d) to print values of each word printed by list(orderset).
  '''
  print("Most popular words: ")
  for i in range(9): #first 10 words in list(orderset) and dict(sort_d)
    print(orderset[i],":",sort_d[orderset[i]]) #print formatting
######################################################
def sort():
  '''
Creates new dict(sort_d) and list(orderset) by ordering dict(wordct) by 
numerical value in descending order.
  '''
  for i in reversed(range(5000)): #i is set to 5,000 and reverses to 0
                                  #i used as count of each word
    for x in wordct: #checks each word in dict(wordct)
      if wordct[x] == i: #if word x appeares i times
        if x != '':
          sort_d[x]=i #adds dict key and value to new dict(sort_d)
          orderset.append(x) #adds the word to list(orderset)
######################################################
def index():
  '''
  Counts copies of words found by urlreader, creates numerical count in
  dict(wordct) for each unique word
  '''
  for word in tempwordct: #looks at each word assigned in list(tempwordct)
    if word in wordct.keys(): #if the word has previously been assigned to dict(wordct)
      wordct[word] += 1
    else: #if the word is new to the dict(wordct)
      wordct[word] = 1
######################################################
def urlreader(url):
  '''
takes input (url) as a website address, then finds all words in website.
Then adds all to a list, then counted and placed into a dictionary with each word count.
New dictionary is made with order of word count in descending order.
  '''
  x = urllib.request.urlopen(url)
  for line in x: #for each line in the url:
    line1 = line.decode('utf-8').strip().lower() #translates byte line into str
    tempword = ''  #resets the temporary word to None
    for c in line1: #for each chacter in the line
      if c not in breakers: #if c is not a character that "breaks" a word
        tempword += c
      elif c in breakers: #if c is a breaking character
        tempwordct.append(tempword) #adds temporary word to list(tempwordct)
        tempword = ''
      if c is line1[-1]: #if c is the final character in the line
        if c not in breakers:
          tempwordct.append(tempword) #adds whatever is in the temp word if the final
                                      #character is not a breaking character
  index()
  sort()
  output()
  

```
