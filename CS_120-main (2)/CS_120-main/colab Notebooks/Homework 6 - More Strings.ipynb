```python
import urllib.request
#creates word bank to check with the text file
corruption = ['decay', 'death', 'poison', 'mould', 'rot','rotten','conceit']
#initializes word count
count=0
#openes website
temp = urllib.request.urlopen('http://library.beau.org/gutenberg/etext00/0ws2610.txt')
#checks each line in the website
for line in temp:
  #each line is formatted into a readable text
  line1 = line.decode('utf-8').strip()
  #checks each line for each word in the word bank
  for word in corruption:
    if word in line1:
      #count increases for each word that appears each time in the file.
      count+=1
      
```
