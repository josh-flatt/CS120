```python
##Question 1

from numpy import sqrt

def bab_q1(s,x):
  '''
  S = the number to be square rooted. X = the guess for the square root.
  The function runs 10 times to be more precisely accurate than just once.
  '''
  for i in range(0,10):
      eps = float((s-(x**2))/((2*x)+0))
      x += eps
        
  return x

def accuracy_q1(s,x):
  '''
  Runs babylonian method until it is accurate to the 10^-4.
  '''
  while (abs(sqrt(s)-bab_q1(s,x)>(10**-4))):
    bab_q1(s,x)
  else:
    return bab_q1(s,x)
```

```python
accuracy_q1(2,5)
```
