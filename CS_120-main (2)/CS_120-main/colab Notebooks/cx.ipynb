```python
'''
Functions for dealing with complex numbers as tuples
'''
```

```python

def cxsum(z1,z2):
  real = z1[0]+z2[0]
  imaginary = z1[1]+z2[1]
  return real,imaginary
```

```python
def cxproduct(z1,z2):
  a = (z1[0]*z2[0])-(z1[1]*z2[1])
  b = (z1[0]*z2[1])+(z2[0]*z1[1])
  return a,b
```

```python
def cxneg(z):
  a = -z[0]
  b = -z[1]
  return a,b
```

```python
def cxpower(z,p):
  temp = [1,1]
  if p == 1:
    return z
  elif p > 1:
    return cxproduct(z,cxpower(z,p-1))
```

```python
def cxmag(z):
    from numpy import sqrt
    return sqrt(z[0]**2 + z[1]**2)
```

```python
def cxdiv(z1,z2):
  r = z2[0]**2 + z2[1]**2
  a = ((z1[0]*z2[0])+(z1[1]*z2[1]))/r
  b = ((z2[0]*z1[1])-(z1[0]*z2[1]))/r
  return a,b
```

```python
def cxe(S, x):
    num = cxsum(S, cxneg(cxpower(x, 2)))
    den = cxproduct( (2, 0), x)
    return cxdiv(num, den)
```

```python
def cxsqrt(S, x):
    count = 0
    er = cxe(S, x)
    while (cxmag(er)> 10**-15) and (count < 20):
        x = cxsum(x, er)
        er = cxe(S, x)
        count += 1
    return x, cxmag(er), count
```
